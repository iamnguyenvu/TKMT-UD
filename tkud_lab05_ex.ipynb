```python
# Import các thư viện thông dụng
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import các distribution packages từ thư viện scipy
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import t
```

```python
# **Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

# a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.
import random

def random_ans(number):
    count = 0
    for _ in range(number):
        if random.choice([0, 0, 0, 1]) == 1:
            count += 1
    print(f"So cau dung: {count}")
for _ in range(10):
    random_ans(10)

# print(f"So cau dung: {binom.rvs(10, 0.25)}")

# b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm
n, p = 10, 0.25
print(f" Xac suat de sinh vien duoc 4 diem: {binom.pmf(4, 10, 0.25)}")

# c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)
print(f"E_X: {n * p}")
print(f"Var_X: {n * p * (1 - p)}")
```

d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

- $P(X \leq 20)$
- $P(X \ge 29.5)$
- $P(X = 23.8)$

```python
ppf_975 = norm.ppf(0.975)
ppf_975_1_1 = norm.ppf(0.975, 1, 1)
ppf_975_1_2 = norm.ppf(0.975, 1, 2)

print(f"norm.ppf(0.975) = {ppf_975:.4f}")
print(f"norm.ppf(0.975, 1, 1) = {ppf_975_1_1:.4f}")
print(f"norm.ppf(0.975, 1, 2) = {ppf_975_1_2:.4f}")

x = np.linspace(-5, 5, 1000)
y = norm.pdf(x)

# Vẽ phân phối chuẩn chuẩn hóa
plt.plot(x, y, label='N(0,1)', color='blue')

# Vẽ phân phối chuẩn với mu=1, sigma=1
y2 = norm.pdf(x, 1, 1)
plt.plot(x, y2, label='N(1,1)', color='green')

# Vẽ phân phối chuẩn với mu=1, sigma=2
y3 = norm.pdf(x, 1, 2)
plt.plot(x, y3, label='N(1,2)', color='red')

# Thêm các điểm phân vị
plt.axvline(ppf_975, color='blue', linestyle='--', label='ppf(0.975) N(0,1)')
plt.axvline(ppf_975_1_1, color='green', linestyle='--', label='ppf(0.975) N(1,1)')
plt.axvline(ppf_975_1_2, color='red', linestyle='--', label='ppf(0.975) N(1,2)')

plt.legend()
plt.title('Minh họa phân phối chuẩn và các điểm phân vị')
plt.show()
```
