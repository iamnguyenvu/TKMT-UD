## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np

arr = np.random.randint(1, 1000, size=100)

print(arr)
print(arr[:10])
print(arr[-5:])

elements = [x for x in arr if x % 5 == 0 and x > 100]
print(elements)
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
smaller = [x for x in arr if x < 100]
print(len(smaller))
print(smaller)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
import numpy as np

print(sum(arr))
print(sum(arr) / len(arr))

median = np.median(arr)
print(median)

Q1 = np.percentile(arr, len(arr) / 4)
Q2 = median
Q3 = np.percentile(arr, len(arr) * 3 / 4 )
print(f"{Q1}")
print(f"{Q2}")
print(f"{Q3}")
print(np.quantile(arr, [0.25, 0.5, 0.75]))

print(np.percentile(arr, 20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
n = np.random.randint(6,10)
print(n)
for i in range(n):
    res = np.random.choice(arr, size=10)
    print(res)
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
# sử dụng np.unique(), hàm này trả về các phần tử duy nhất trong mảng,
# nếu return_counts=True còn trả về số lần xuất hiện của từng phần tử

unique_elements, counts = np.unique(arr, return_counts=True)
# unique_elements là mảng chứa các phần tử duy nhất trong mảng
# counts là mảng chứa só lần xuất hiện của từng phần tử trong unique_elements

# np.argmax() trả về chỉ số của phần tử có giá trị lớn nhất trong mảng
print(unique_elements[np.argmax(counts)])
print(counts[np.argmax(counts)])

print(f'Phần tử xuất hiện nhiều nhất: {unique_elements[np.argmax(counts)]}')
print(f'Số lần xuất hiện: {counts[np.argmax(counts)]}')
```

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
plt.subplot(2, 1, 1)
x = np.random.normal(size=10000)
plt.ylabel('Mật dộ')
plt.title('Histogram')
plt.hist(x, bins=100, color='purple', density=True)

plt.subplot(2, 1, 2)
plt.boxplot(x, vert=False)
plt.title('Boxplot')

plt.show()
```
